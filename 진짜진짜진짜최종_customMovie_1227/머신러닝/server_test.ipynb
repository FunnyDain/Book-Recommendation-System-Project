```python
from flask import Flask, request
from flask_cors import CORS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
app = Flask(__name__)
CORS(app)


@app.route('/EN', methods=['GET','POST'])
def EN():
    if request.method == 'GET':
        data = request.args.get("movie")
        df = pd.read_csv('영화리스트.csv',encoding='euc-kr',header= 0)
        df.dropna(axis=1)
        test = df[['E/I','S/N','T/F','J/P']]
        X = test.iloc[:, :].values
        scaler = StandardScaler()
        data_scaled = scaler.fit_transform(test)
        pca = PCA(n_components=2)
        pca.fit(data_scaled)
        df_pca = pca.transform(data_scaled)
        K = 4
        kmeans = KMeans(n_clusters=K, random_state=0)
        y_kmeans = kmeans.fit_predict(df_pca)
        model = kmeans.fit(df_pca)
        df_X_org = pd.DataFrame(X)
        result_df = pd.concat([df, pd.Series(y_kmeans)], axis=1)
        mbti_movie = result_df
        mbti_movie = mbti_movie.dropna(axis=1)
        EN = mbti_movie[result_df.iloc[:, -1] == 2]
        EN = EN.query("장르 in ['로맨스','드라마','액션']")[:8]
        EN = EN[['영화제목']]
        return EN.to_json(force_ascii=False)

@app.route('/ES', methods=['GET','POST'])
def ES():
    if request.method == 'GET':
        df = pd.read_csv('영화리스트.csv',encoding='euc-kr',header= 0)
        df.dropna(axis=1)
        test = df[['E/I','S/N','T/F','J/P']]
        X = test.iloc[:, :].values
        scaler = StandardScaler()
        data_scaled = scaler.fit_transform(test)
        pca = PCA(n_components=2)
        pca.fit(data_scaled)
        df_pca = pca.transform(data_scaled)
        K = 4
        kmeans = KMeans(n_clusters=K, random_state=0)
        y_kmeans = kmeans.fit_predict(df_pca)
        model = kmeans.fit(df_pca)
        df_X_org = pd.DataFrame(X)
        result_df = pd.concat([df, pd.Series(y_kmeans)], axis=1)
        mbti_movie = result_df
        mbti_movie = mbti_movie.dropna(axis=1)
        ES = mbti_movie[result_df.iloc[:, -1] == 1]
        ES = ES.query("장르 in ['로맨스','드라마','스릴러','액션']")[:8]
        ES = ES[['영화제목']]
        return ES.to_json(force_ascii=False)

@app.route('/IN', methods=['GET','POST'])
def IN():
    if request.method == 'GET':
        df = pd.read_csv('영화리스트.csv',encoding='euc-kr',header= 0)
        df.dropna(axis=1)
        test = df[['E/I','S/N','T/F','J/P']]
        X = test.iloc[:, :].values
        scaler = StandardScaler()
        data_scaled = scaler.fit_transform(test)
        pca = PCA(n_components=2)
        pca.fit(data_scaled)
        df_pca = pca.transform(data_scaled)
        K = 4
        kmeans = KMeans(n_clusters=K, random_state=0)
        y_kmeans = kmeans.fit_predict(df_pca)
        model = kmeans.fit(df_pca)
        df_X_org = pd.DataFrame(X)
        result_df = pd.concat([df, pd.Series(y_kmeans)], axis=1)
        mbti_movie = result_df
        mbti_movie = mbti_movie.dropna(axis=1)
        IN = mbti_movie[result_df.iloc[:, -1] == 0]
        IN = IN.query("장르 in ['드라마','코미디','재난']")[:8]
        IN = IN[['영화제목']]
        return IN.to_json(force_ascii=False)

@app.route('/IS', methods=['GET','POST'])
def IS():
    if request.method == 'GET':
        df = pd.read_csv('영화리스트.csv',encoding='euc-kr',header= 0)
        df.dropna(axis=1)
        test = df[['E/I','S/N','T/F','J/P']]
        X = test.iloc[:, :].values
        scaler = StandardScaler()
        data_scaled = scaler.fit_transform(test)
        pca = PCA(n_components=2)
        pca.fit(data_scaled)
        df_pca = pca.transform(data_scaled)
        K = 4
        kmeans = KMeans(n_clusters=K, random_state=0)
        y_kmeans = kmeans.fit_predict(df_pca)
        model = kmeans.fit(df_pca)
        df_X_org = pd.DataFrame(X)
        result_df = pd.concat([df, pd.Series(y_kmeans)], axis=1)
        mbti_movie = result_df
        mbti_movie = mbti_movie.dropna(axis=1)
        IS = mbti_movie[result_df.iloc[:, -1] == 3]
        IS = IS.query("장르 in ['스릴러','범죄']")[:8]
        IS = IS[['영화제목']]
        return IS.to_json(force_ascii=False)

@app.route('/ranking', methods=['GET','POST'])
def ranking():
    df = pd.read_csv('영화리스트.csv',encoding='euc-kr',header= 0)
    df.dropna(axis=1)
    def weight_vote_average(record):
        m = df['vote_count'].quantile(0.9)
        C = df['vote_average'].mean()
        v = record['vote_count']
        R = record['vote_average']
        return ((v/(v+m)) * R ) +  ((m/(m+v)) * C)
    def ranking():
        df['score'] = df.apply(weight_vote_average, axis=1)
        df1 = df[['영화제목','score']].sort_values('score',ascending=False)[:9]

        return df1[['영화제목']]

    return ranking().to_json(force_ascii=False)




app.run(host="0.0.0.0",port=5000)
```
